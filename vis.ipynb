## SCATTER PLOT

### Base Example

```
x = np.array([1,2,3,4,5,6,7,8])
y = x

plt.figure()
plt.scatter(x,y,)
plt.title("Simple scatterplot")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")

```

### More params

```
observations = 50
gradient = 2
x = np.array(range(observations))
y = np.random.randn(observations) * 10 + x*gradient

fig, ax = plt.subplots()
colors = np.random.rand(observations) 
ax.scatter(x, y,
           s=5,# size
           c=colors,# colors according to array c
           alpha=0.5 # 0 = transparent, opaque = 1
           ) 
fig.suptitle('Scatterplot parameters (color,size and alpha')
```

### Using Pandas

```
# color depends on col C
d = {'A': np.arange(10),'B': np.arange(10), 'C': [1,2,1,2,1,2,1,2,1,2]}
df = pd.DataFrame(data=d)
df.plot.scatter('A',
                'B', 
                c='C', 
                colormap='viridis',
                title='Scatter plot using Pandas')
```

# Bar Charts

## Base example

```
plt.figure()
x = list(range(1,9))         # define categories
y = [0,0,1,2,3,3,2,0] # how many per category


print(x)
print(y)
```

```
plt.bar(x, y, width = 1, color='green')
plt.ylabel('Y ')
plt.title("Barchart example")

plt.tick_params(top='off', bottom='off', left='off', right='off', labelleft='off', labelbottom='off')
```

## Bar Stack

```
y2 = [el / 2 for el in y] # how many per category
plt.figure()
plt.bar(x, y,  width = 1, color='green')
plt.bar(x, y2, width = 1, color='red',bottom=y2)
```
